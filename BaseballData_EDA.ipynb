# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# 1. Import the database as a Pandas dataframe.

```python
file = "Baseball Data.xlsx"
data = pd.read_excel(file)
data
```

## a. Which player had the highest batting average? The lowest? 

### Player with highest batting average:

```python
data[data['Batting Average'] == data['Batting Average'].max()]
```

### Player(s) with lowest batting average:

```python
data[data['Batting Average'] == data['Batting Average'].min()]
```

## b. Which player had the highest # of home-runs per game?

```python
data['Home-runs per Game'] = data['Home Runs'] / data['Games Played']
data
```

```python
data[data['Home-runs per Game'] == data['Home-runs per Game'].max()]
```

## c. Do players with higher batting averages tend to score more home runs per game? 

```python
data.plot.scatter(x='Batting Average', y = 'Home-runs per Game')
```

There does not seem to be a correlation between batting average scores and average home-runs per game. 

# 2. Statistical Analysis

```python
gamesPlayed = data['Games Played'].to_numpy()
gamesPlayed
```

## a. What is the mean number of games played? The median?

```python
mean = gamesPlayed.mean()
mean
```

```python
median = np.median(gamesPlayed)
median
```

## b. Histogram

```python
plt.hist(gamesPlayed, bins = 15)
```

Observing the histogram, data seems to be clustering at the tails rather than the at the average.

# 3. Isolate players who play on 1st Base, and those who play outfield.

```python
first_base = data.loc[(data['POS'] == '1B')]
first_base
```

```python
outfield = data.loc[(data['POS'] == 'OF')]
outfield
```

## a. To which position does the highest home-run scoring player belong to?

```python
first_base[first_base['Home Runs'] == first_base['Home Runs'].max()]
```

```python
outfield[outfield['Home Runs'] == outfield['Home Runs'].max()]
```

The highest home-run scoring player is Chris Carter who is first baser.

## b. Compare the means and medians of batting averages. Can you conclude that one group hits more successfully than the other?

```python
firstBase_BA_mean = first_base['Batting Average'].to_numpy().mean()
firstBase_BA_mean
```

```python
firstBase_BA_median = np.median(first_base['Batting Average'].to_numpy())
firstBase_BA_median
```

```python
outfield_BA = outfield['Batting Average'].to_numpy().mean()
outfield_BA
```

```python
outfield_BA_median = np.median(outfield['Batting Average'].to_numpy())
outfield_BA_median
```

```python
stats.ttest_ind(a=first_base['Batting Average'].to_numpy(), b=outfield['Batting Average'].to_numpy(), equal_var=True)
```

We cannot conclude that one position has better batting averages than the other.
